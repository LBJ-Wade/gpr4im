## Loading in data and Fourier transforming

Also calculating frequency covariance matrix in Fourier space

```python
# loading image data:
HI_data = np.load('../Data/dT_HI_smoothed.npy')
HI_data_nb = np.load('../Data/dT_HI.npy') # nb = no beam
FG_nopol_data = np.load('../Data/dT_FGnopol_Stripe82_smoothed.npy')
FG_wpol_data = np.load('../Data/dT_FGwpol_Stripe82_smoothed.npy')
FG_nopol_data_nb = np.load('../Data/dT_FGnopol_Stripe82.npy')
FG_wpol_data_nb = np.load('../Data/dT_FGwpol_Stripe82.npy')

# converting to visbility:
df = 1 # frequency resolution of data (MHz)
vmin = 899 # min frequency (MHz)
vmax = 1184 # max frequency (MHz)
nx = np.shape(HI_data)[0]
ny = np.shape(HI_data)[1] # number of pixels in nx and ny direction of data (equal in our case)
ra_width = dec_width = 54.1 # width of sky in RA and Dec directions (deg)
HI_data_vis = load.generateCartDataCube(HI_data, vmin, vmax, df, nx, ny, ra_width, dec_width)
HI_data_nb_vis = load.generateCartDataCube(HI_data_nb, vmin, vmax, df, nx, ny, ra_width, dec_width)
FG_nopol_data_vis = load.generateCartDataCube(FG_nopol_data, vmin, vmax, df, nx, ny, ra_width, dec_width)
FG_nopol_data_nb_vis = load.generateCartDataCube(FG_nopol_data_nb, vmin, vmax, df, nx, ny, ra_width, dec_width)
FG_wpol_data_vis = load.generateCartDataCube(FG_wpol_data, vmin, vmax, df, nx, ny, ra_width, dec_width)
FG_wpol_data_nb_vis = load.generateCartDataCube(FG_wpol_data_nb, vmin, vmax, df, nx, ny, ra_width, dec_width)
```

Calculate frequency covariance matrices:

```python
# visibility data:
C_HI_vis = np.cov(HI_data_vis.data)
C_HI_nb_vis = np.cov(HI_data_nb_vis.data)
C_FGnopol_vis = np.cov(FG_nopol_data_vis.data[::-1]) # [::-1] inverts frequency axis
C_FGnopol_nb_vis = np.cov(FG_nopol_data_nb_vis.data[::-1])
C_FGwpol_vis = np.cov(FG_wpol_data_vis.data[::-1])
C_FGwpol_nb_vis = np.cov(FG_wpol_data_nb_vis.data[::-1])

# image data:
C_HI = np.cov(load.LoSpixels(HI_data))
C_HI_nb = np.cov(load.LoSpixels(HI_data_nb))
C_FGnopol = np.cov(load.LoSpixels(FG_nopol_data)[::-1])
C_FGnopol_nb = np.cov(load.LoSpixels(FG_nopol_data_nb)[::-1])
C_FGwpol = np.cov(load.LoSpixels(FG_wpol_data)[::-1])
C_FGwpol_nb = np.cov(load.LoSpixels(FG_wpol_data_nb)[::-1])
```

## HI Data

```python
X = HI_data_vis.freqs/1e6 # frequency range in MHz
kernel_name='Exponential'
# get Z (samples from multivariate distribution with mean zero and covariance C) and C (covariance):
Z, C = cov.get_kernel_samples(kernel_name, X, 20, l=2)
```

```python
fig = plt.figure(figsize=(19,19))     # open a new plotting window
plt.subplot(331)
for i in range(2):
    plt.plot(X[:],Z[i,:])
plt.title('{} samples'.format(kernel_name))

plt.subplot(332)
pos = plt.imshow(C, interpolation='nearest')
fig.colorbar(pos, fraction=0.045)
plt.title('{} covariance'.format(kernel_name))

plt.subplot(334)
for i in (11,110,210):
    plt.plot(X, np.real(load.LoSpixels(HI_data)[:,i]))
plt.title('HI data samples, w beam')

plt.subplot(335)
for i in (11,110,210):
    plt.plot(X, np.real(HI_data_vis.data[:,i]))
plt.title('HI data samples, w beam (VIS)')

plt.subplot(336)
pos = plt.imshow(np.real(C_HI), interpolation='nearest')
fig.colorbar(pos, fraction=0.045)
plt.title('HI data covariance, w beam')

plt.subplot(337)
for i in (11,110,210):
    plt.plot(X, np.real(load.LoSpixels(HI_data_nb, mean_center=False)[:,i]))
plt.title('HI data samples, no beam')

plt.subplot(338)
for i in (11,110,210):
    plt.plot(X, np.real(HI_data_nb_vis.data[:,i]))
plt.title('HI data samples, no beam (VIS)')

plt.subplot(339)
pos = plt.imshow(np.real(C_HI_nb), interpolation='nearest')
fig.colorbar(pos, fraction=0.045)
plt.title('HI data covariance, no beam')
```

## Isolating Beam

```python
# frequency covariance of beam, by dividing data with beam by data without beam in visibility space
#C_fg_b = np.cov((FG_nopol_data_vis.data/FG_nopol_data_nb_vis.data)[::-1])
#C_hi_b = np.cov((HI_data_vis.data/HI_data_nb_vis.data)[::-1])

# frequency covariance of beam, by dividing data with beam by data without beam in real space
C_fg_b = np.cov((load.LoSpixels(FG_nopol_data)/load.LoSpixels(FG_nopol_data_nb))[::-1])
C_hi_b = np.cov((load.LoSpixels(HI_data)/load.LoSpixels(HI_data_nb))[::-1])
```

```python
fig = plt.figure(figsize=(12,12))

plt.subplot(221)
for i in (11, 100, 200):
    plt.plot(X, (load.LoSpixels(FG_nopol_data)[:,i][::-1])/(load.LoSpixels(FG_nopol_data_nb)[:,i][::-1]))
    #plt.plot(X, np.real(FG_nopol_data_vis.data/FG_nopol_data_nb_vis.data)[:,i][::-1])
plt.title('FG data beam / no beam LoS samples')

plt.subplot(222)
for i in (11, 100, 200):
    plt.plot(X, (load.LoSpixels(HI_data)[:,i][::-1])/(load.LoSpixels(HI_data_nb)[:,i][::-1]))
    #plt.plot(X, np.real(HI_data_vis.data/HI_data_nb_vis.data)[:,i][::-1])
plt.title('HI data beam / no beam LoS samples')

plt.subplot(223)
pos = plt.imshow(np.log10(np.real(C_fg_b)), interpolation='nearest')
fig.colorbar(pos, fraction=0.045)
plt.title('Log beam covariance FG (no pol)')

plt.subplot(224)
pos = plt.imshow(np.log10(np.real(C_hi_b)), interpolation='nearest')
fig.colorbar(pos, fraction=0.045)
plt.title('Log beam covariance HI')
```
